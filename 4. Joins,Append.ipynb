```python
import pandas as pd
import numpy as np

# data frame 1
d1 = {'Customer_id' : pd.Series([1,2,3,4,5,6]),
     'Product':pd.Series(['Oven','Oven','Oven',
                         'Television',
                         'Television',
                         'Television'])}
df1 = pd.DataFrame(d1)

#data frame 2
d2 = {'Customer_id':pd.Series([2,4,6,8]),
     'State':pd.Series(['California','California','Texas','Texas'])}
df2 = pd.DataFrame(d2)
```

# Inner join pandas

Returnas all row from both tables, join records from the left which have matching

```python
# inner join in python pandas
pd.merge(df1,df2,on = 'Customer_id',how = 'inner')
```

# Outer join pandas

Returnas all row from both tables, join records from the left which have matching keys in the right table

```python
# Outer join in python pandas
pd.merge(df1, df2, on = 'Customer_id',how = 'outer')
```

# Left Outer join pandas

Returnas all row from the left tables, and any row with matching keys from the right table

```python
# Left join in python pandas
pd.merge(df1, df2, on = 'Customer_id',how = 'left')
```

#  Right Outer join pandas

Returnas all row from the right tables, and any row with matching keys from the left table

```python
# Right join in python pandas
pd.merge(df1, df2, on = 'Customer_id',how = 'right')
```
